```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
epsilon_values = [0.5, 1.0, 1.5, 2.0, 3.0]

# Car Insurance Data Accuracy
nn_accuracy_insurance = [93.55, 93.55, 93.55, 93.55, 93.55]
knn_accuracy_insurance = [93.55, 79.22, 92.69, 92.52, 93.35]
xgb_accuracy_insurance = [93.55, 93.55, 93.55, 93.55, 93.55]

# Diabetes Prediction Data Accuracy
nn_accuracy_diabetes = [68.83, 64.29, 64.94, 64.94, 64.29]
knn_accuracy_diabetes = [35.71, 55.84, 60.39, 52.60, 69.48]
xgb_accuracy_diabetes = [58.44, 58.44, 59.74, 63.64, 68.18]

# Loan Default Data Accuracy
nn_accuracy_loan = [90.97, 90.97, 90.97, 90.97, 90.97]
knn_accuracy_loan = [91.00, 90.00, 90.00, 89.00, 90.00]
xgb_accuracy_loan = [91.00, 91.00, 91.00, 91.00, 91.00]


fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plotting for Car Insurance Data
axes[0].plot(epsilon_values, nn_accuracy_insurance, label='Neural Network', marker='o', linestyle='--', linewidth=2)
axes[0].plot(epsilon_values, knn_accuracy_insurance, label='KNN', marker='o')
axes[0].plot(epsilon_values, xgb_accuracy_insurance, label='XGBoost', marker='o')
axes[0].set_title('Car Insurance Data - Model Accuracy vs. Epsilon')
axes[0].set_xlabel('Epsilon')
axes[0].set_ylabel('Accuracy (%)')
axes[0].legend()

# Plotting for Diabetes Prediction Data
axes[1].plot(epsilon_values, nn_accuracy_diabetes, label='Neural Network', marker='o')
axes[1].plot(epsilon_values, knn_accuracy_diabetes, label='KNN', marker='o')
axes[1].plot(epsilon_values, xgb_accuracy_diabetes, label='XGBoost', marker='o')
axes[1].set_title('Diabetes Prediction Data - Model Accuracy vs. Epsilon')
axes[1].set_xlabel('Epsilon')
axes[1].set_ylabel('Accuracy (%)')
axes[1].legend()

# Plotting for Loan Default Data
axes[2].plot(epsilon_values, nn_accuracy_loan, label='Neural Network', marker='o')
axes[2].plot(epsilon_values, knn_accuracy_loan, label='KNN', marker='o')
axes[2].plot(epsilon_values, xgb_accuracy_loan, label='XGBoost', marker='o')
axes[2].set_title('Loan Default Data - Model Accuracy vs. Epsilon')
axes[2].set_xlabel('Epsilon')
axes[2].set_ylabel('Accuracy (%)')
axes[2].legend()


# Adjust layout
plt.tight_layout()
plt.show()

```
